```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
Cars_Data = pd.read_csv(r'C:\Users\Nikita\OneDrive\Desktop\Python for DataScience\dataset\UAE_USED_CARS_DATASET.csv')
df = pd.DataFrame(Cars_Data)
```

```python
# Calculate Mean And Standard Daviation 
mean = np.mean(df['Price'])
std = np.std(df['Price'])
# Calculate Z-Score
df['Z-Score']=(df['Price']-mean)/std
# Print the data
print("-------------------------------------------------------------------")
print(f" Here is the data with outliers: \n {df}")
print("-------------------------------------------------------------------")
# Print The Outliers
print(f" Here are the outliers based on the z-score threshold , 3: \n {df[df['Z-Score']>3]}")
# Remove the outliers
df = df[df['Z-Score']<=3]
#Print The Data without outliers
print(f"Here is the data without outliers: \n{df}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histogram to visualize skewness
plt.figure(figsize=(10,6))
sns.histplot(df['Price'], kde=True, bins=30, color='skyblue')
plt.title('Histogram of Price (After Removing Outliers)', fontsize=16)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
df['Log_Price'] = np.log1p(df['Price'])  # Using log1p to handle zero values
plt.figure(figsize=(10, 6))
sns.histplot(df['Log_Price'], kde=True, bins=30, color='skyblue')
plt.title('Histogram of Log-Transformed Price', fontsize=16)
plt.xlabel('Log(Price)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
