```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
Cars_Data = pd.read_csv(r'C:\Users\Nikita\OneDrive\Desktop\Python for DataScience\dataset\UAE_USED_CARS_DATASET.csv')

# Select numeric columns
numeric_data = Cars_Data.select_dtypes(include=['number'])

# Plot histograms for skewness visualization
for column in numeric_data.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(numeric_data[column], kde=True, color='blue', bins=30)
    plt.title(f'Distribution of {column} (Skewness: {numeric_data[column].skew():.2f})')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

# Alternatively, plot box plots for numeric columns
plt.figure(figsize=(10, 6))
sns.boxplot(data=numeric_data, orient="h")
plt.title('Box Plot for Numeric Columns')
plt.xlabel('Values')
plt.show()
```

#Here’s an analysis of your second histogram titled "Distribution of Price (Skewness: 9.28)":Variable Analyzed: Price
📈 Distribution Summary:
Skewness: 9.28 → This is a highly positively skewed (right-skewed) distribution.

Visual Pattern:

Most of the values are clustered on the left (low price range).

A long tail stretches far to the right, with fewer but extreme high-price values.

The x-axis reaches up to 14 million+, but most data lies well below 2 million.

🧠 Interpretation:
A few very high prices are pulling the mean to the right — these are likely outliers or luxury-priced items.

The median price would be much lower than the mean.

This pattern is typical in real-world data where:

Most products/services are affordable

A few are extremely expensive (e.g., luxury houses, cars, high-end goods)

🔍 Insights:
Using mean alone would be misleading — use median for central tendency.

Many models (like linear regression) could perform poorly due to the skew.
